## FINE TUNING DE MODELOS: RF, LIGHTGBM Y XGB

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_distributions = {
    'n_estimators': [50, 100, 150, 400, 600],  
    'max_depth': [None, 10, 20, 30],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4], 
    'bootstrap': [True, False], 
    'max_features': ['auto', 'sqrt', 'log2']
    }
```

```python
random_search_rf = RandomizedSearchCV(estimator=rf_reg, param_distributions=param_distributions, 
                                      n_iter=14, cv=5, verbose=2, random_state=1312)
```

```python
random_search_rf.fit(X_train, y_train)
```

```python
best_params = random_search_rf.best_params_
```

```python
best_params
```

```python
mod_rf = RandomForestRegressor(**best_params, random_state=1312, verbose=4)
```

```python
mod_rf.fit(X_train, y_train)
```

```python
mod_rf
```

```python
y_pred = mod_rf.predict(X_test)
```

```python
param_grid_lgbm = {
    'num_leaves': [30, 63, 200, 300, 400, 600],  # Valores pequeños para limitar la complejidad del modelo
    'max_depth': [-1, 7, 15, 20, 30],  # Profundidad máxima del árbol (-1 significa sin límite, prof total)
    'learning_rate': [0.01, 0.05, 0.1, 0.5, 0.3],  # Tasa de aprendizaje, peso de los predictores anteriores
    'n_estimators': [50, 100, 200, 300, 500, 600],  # Número de árboles para cada estimación
    'min_child_samples': [20, 50],  # Número mínimo de datos en una hoja
    'subsample': [0.8, 1.0],  # Proporción de datos utilizados en cada iteración
    'colsample_bytree': [0.8, 1.0]  # Proporción de características utilizadas en cada iteración

}
```

```python
lgbm = lgb.LGBMRegressor(random_state=42, max_depth= -1 )
```

```python
random_search_lgbm = RandomizedSearchCV(estimator=lgbm, param_distributions= param_grid_lgbm, 
                                         n_iter=80, cv=5, verbose=4, random_state=42)
```

```python
random_search_lgbm.fit(X_train, y_train)
```

```python
best_params_lgbm = random_search_lgbm.best_params_
```

```python
best_params_lgbm
```

```python
modelolgbm = lgb.LGBMRegressor(**best_params, random_state=1213, verbose_eval=4)
modelolgbm.fit(X_train, y_train)
```

```python
y_pred_lgbm = modelolgbm.predict(X_test)
```

```python
mse = root_mean_squared_error(y_test, y_pred_lgbm)
```

```python
mod_lgbm = 'modelolgbm.pkl'

with open(mod_lgbm, 'wb') as file:
    pickle.dump(modelolgbm, file)

print(f'Modelo guardado bien')
```
