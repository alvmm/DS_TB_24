## Regresion Lineal Simple OLS sm

```python
X_train = df_clean[['duration_ms', 'explicit', 'danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo', 'time_signature']]

X_train = sm.add_constant(X_train, prepend=True)

modelo = sm.OLS(endog=df_clean["popularity"], exog=X_train)

modelo = modelo.fit()

print(modelo.summary())
```

```python
# 0.025 de R2 jejej         #Es normal este valor taaaan bajo?
```
